# Handling Imbalance Data with SMOTE

Synthetic Minority Over-sampling Technique

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res, y_res = SMOTE().fit_resample(x,y) 
```

```python
x_res.value_counts()
```

```python
y_res.value_counts()
```

# Model Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_res,y_res, test_size=0.20, random_state=42)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
y_pred1 = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred1)
```

```python
accuracy_score(y_test, y_pred1)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(y_test, y_pred1)
```

```python
recall_score(y_test, y_pred1)
```

```python
f1_score(y_test, y_pred1)
```

# Visualization

```python
import matplotlib.pyplot as plt

print(data['Exited'].dtype)
print(data['Balance'].dtype)


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_balance = scaler.fit_transform(data[['Balance']])

plt.scatter(scaled_balance, data['Exited'])
plt.xlabel('Scaled Balance')
plt.ylabel('Exited')
plt.title('Scaled Balance vs. Exited')
plt.show()
```

# Barplot

```python
# Box plot for 'Age' grouped by 'Exited'
sns.boxplot(x='Exited', y='Age', data=data)
plt.xlabel('Exited')
plt.ylabel('Age')
plt.title('Age Distribution by Exited')
plt.show()
```

```python
# 2. Balance vs exited
sns.boxplot(x='Exited', y='Balance', data=data)
plt.xlabel('Exited')
plt.ylabel('Balance')
plt.title('Balance Distribution by Exited')
plt.show()
```

```python
# 3. Credit score vs exited
sns.boxplot(x='Exited', y='CreditScore', data=data)
plt.xlabel('Exited')
plt.ylabel('Credit Score')
plt.title('Credit Score Distribution by Exited')
plt.show()
```

```python
#4. Estimated salary vs Exited
sns.boxplot(x='Exited', y='EstimatedSalary', data=data)
plt.xlabel('Exited')
plt.ylabel('Estimated Salary')
plt.title('Estimated Salary Distribution by Exited')
plt.show()
```

# Correlation Matrix

```python
# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

1. Correlation matrix with heatmap

```python
selected_features = ['Age', 'Balance', 'CreditScore', 'EstimatedSalary']
selected_correlation_matrix = data[selected_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(selected_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Selected Features')
plt.show()
```

2. Correlation matrix with pairplot

```python
sns.pairplot(data[selected_features])
plt.title('Pairplot for Selected Features')
plt.show()
```
